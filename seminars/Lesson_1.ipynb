Получим статистические характеристики параметров с помощью numpy:

```python
# Чтение имен параметров из первой строки
with open('winequality-red.csv') as infile:
	names = infile.readline().rstrip('\n').split(';')

# Чтение данных без первой строки
data = np.loadtxt('winequality-red.csv', skiprows = 1, delimiter=';')

# Вывод параметров и их характеристик
for n in range(0, len(names)):
	print('\n' + names[n])
	print('\tmin\t{:.3f}'.format(np.min(data[:, n])))
	print('\tmean\t{:.3f}'.format(np.mean(data[:, n])))
	print('\tmax\t{:.3f}'.format(np.max(data[:, n])))
	print('\tStDev\t{:.3f}'.format(np.std(data[:, n])))
```

- Медиана, как и среднее значение, нужна для определения типичного значения.

- Чтобы найти медиану, данные нужно расположить в порядке возрастания. Медианой будет значение, которое совпадает с серединой набора данных. Если количество значений чётное, то берётся среднее двух значений, которые «окружают» середину.

```python
# медианное значение: это число, которое является серединой множества чисел, 
# то есть половина чисел имеют значения большие, чем медиана, а половина чисел имеют значения меньшие, чем медиана.
np.median(l)
```

Запомните на всю жизнь: выбросы - зло. Среднее значение чувствительно к выбросам, а медиана нет.

```python
# Мода — это число, наиболее часто встречающееся в данном наборе чисел. 
from scipy import stats
int(stats.mode(l)[0])
```

Предельная ошибка выборки
-----

```python
Image('errors.png')
```

- Ошибка выборки – разность между средним значением показателя в выборочной совокупности и среднем значением показателя в генеральной совокупности.
- Предельная ошибка выборки – максимально возможное расхождение значений выборки и генеральной совокупности с заданной нами вероятностью (p=95% ~ t=1.96 ). 


```python
# Размерность таблицы
data_wine.shape
```

```python
# Генерация набора индексов для выборки (возьмем размер выборки = 160)
sample = np.random.randint(1599, size = 160)
sample
```

```python
# Создание выборки
data_wine['total sulfur dioxide'][sample]
```

```python
# Среднее по выборке
data_wine['total sulfur dioxide'][sample].mean()
```

```python
# Расчёт предельной ошибки выборки

# коэффициент доверия Стьюдента (p=95%)
tS = 1.96 

# дисперсия
std = data_wine['total sulfur dioxide'][sample].std() 

# объем генеральной совокупности
N_g = 1599

# необходимая численность выборки
n_s = 160

# предельная ошибка выборки
Dx = tS * np.sqrt( std**2 * (1 - n_s/N_g) / n_s)
Dx
```

Доверительный интервал: 
- Среднее по выборке - предельная ошибка <= X <= Среднее по выборке + предельная ошибка:

```python
print((data_wine['total sulfur dioxide'][sample].mean()-Dx), ' <= X <= ', \
      data_wine['total sulfur dioxide'][sample].mean()+Dx)
```

```python
# Среднее по генеральной совокупности
print ('X = ', data_wine['total sulfur dioxide'].mean())
```

```python
# Дисперсия из pandas
std**2
```

```python
# Ручной расчёт дисперсии
Xi = data_wine['total sulfur dioxide'][sample]
Xm = data_wine['total sulfur dioxide'][sample].mean()
S2 = np.sum(np.power(Xi - Xm, 2))/(n_s - 1)
S2
```

Объём выборки
-----

Генеральная совокупность очень хорошо описывает весь набор данных. И по ней хорошо бы оценить всякие показатели. Однако если у нас нет возможности оценить всю генеральную совокупность, например, мы хотим узнать средний возраст человека на Земле. Да, мы знаем генеральную совокупность — 7 миллиардов людей, но как мы можем узнать средний возраст? То есть опрашивать каждого — это не вариант, потому что, во-первых, это очень много людей, а во-вторых, кто-то умирает, кто-то рождается, возраст меняется, все меняется. Поэтому необходимо взять какой-то небольшой объем данных и по нему оценить. А какой объем данных взять? Насколько он большой или небольшой должен быть?

```python
# p=95%
tS = 1.96

# p=99%
# tS = 2.58

# p=99,9%
# tS = 3.29
```

```python
# Задаём предельную ошибку dx и находим необходимый объём выборки
dx = Dx
# dx = Dx/2
# dx = Dx/4
V_sample = tS**2 * std**2 * N_g / (dx**2 * N_g + tS**2 * std**2)
V_sample
```

```python
data_wine['total sulfur dioxide'].mean()
```

```python
data_wine['total sulfur dioxide'].sample(n = int(V_sample)).mean()
```
